### Gym interface

The three main methods of an environment are
* __reset()__ - reset environment to initial state, _return first observation_
* __render()__ - show current environment state (a more colorful version :) )
* __step(a)__ - commit action __a__ and return (new observation, reward, is done, info)
 * _new observation_ - an observation right after commiting the action __a__
 * _reward_ - a number representing your reward for commiting action __a__
 * _is done_ - True if the MDP has just finished, False if still in progress
 * _info_ - some auxilary stuff about what just happened. Ignore it for now

```python
def run_episode(env, policy, gamma = 1.0, render = False):
    obs = env.reset()
    total_reward = 0
    step_idx = 0
    while True:
        if render:
            env.render()
        obs, reward, done , _ = env.step(int(policy[obs]))
        total_reward += ((gamma ** step_idx) * reward)
        step_idx += 1
        if done:
            break
    return total_reward
```

```python
def evaluate_policy(env, policy, gamma = 1.0,  n = 100):
    scores = [
            run_episode(env, policy, gamma = gamma, render = False)
            for _ in range(n)]
    return np.mean(scores)
```

```python
# Используя фиксированную value_function и жадную стратегию получите policy

def extract_policy(env, v, gamma = 1.0):
    policy = np.zeros(env.env.nS)
    for s in range(env.env.nS):
        best_move = -1
        best_result = -np.inf
        for a in range(env.env.nA):
            probs, states, rews, done = map(list, zip(*env.env.P[s][a]))
            challenger = 0
            for i in range(len(probs)):
                challenger += rews[i] + probs[i] * gamma*v[states[i]]
            if challenger > best_result:
                best_move = a
                best_result = challenger
        policy[s] = best_move
                
    return policy
```

```python
def compute_policy_v(env, policy, v, gamma=1.0):
    for s in range(env.env.nS):
        probs, states, rews, done = map(list, zip(*env.env.P[s][policy[s]]))
        result = 0
        for i in range(len(probs)):
            result += rews[i] + probs[i] *gamma*v[states[i]]
        v[s] = result
    return v
```

```python
# Используя фиксированную policy с помощью сэмплирования получите оценку для value_function с точностью 1e-10
def compute_policy_v_sample(env, policy, v, gamma=1.0):
    for s in range(env.env.nS):
        probs, states, rews, done = map(list, zip(*env.env.P[s][policy[s]]))
        N = 10000
        result = 0
        for _ in range(N):
            i = np.random.choice(len(probs), p=probs)
            rew = rews[i]
            result += (rew + gamma*v[states[i]]) / N
        v[s] = result 
    return v
```

```python
def policy_iteration(env, gamma = 1.0):
    policy = np.zeros(env.env.nS)
    old_v = np.zeros(env.env.nS)
    max_iterations = 10000
    gamma = 1.0
    for i in range(max_iterations):
        old_v = compute_policy_v(env, policy, old_v, gamma)
        new_policy = extract_policy(env, old_v, gamma)

        policy = new_policy
        print(policy)
    return policy
```

```python
gamma = 1.0
env = gym.make("FrozenLake-v0")
optimal_policy = policy_iteration(env, gamma=gamma)
policy_score = evaluate_policy(env, optimal_policy, gamma, n=100)

print(policy_score)
```
